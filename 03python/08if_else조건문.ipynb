# if else 조건문

## 1. if 조건문 : 특정 조건을 판별해서 작업을 분기시키는 문법
 ``` python
if 조건식:
    조건이 참일 때 실행할 코드
else:
    조건이 거짓일 때 실행할 코드
```

````python
x=12
````

````python
#x가 10인지 아닌지 판별하는 조건문
if x==10:
    print(f"{x}는 10입니다")
else:
    print(f"{x}는 10이 아닙니다")    
    
````

````python
#<x가 10 또는 짝수>인지 아닌지 판별하는 조건문
if x==10 or x%2==0:
    print(f"{x}는 10 또는 짝수입니다")
else:
    print(f"{x}는 짝수가 아닙니다")    
````

- 숫자를 입력 받아 홀수인지 짝수인지 판단하는 프로그램

````python
x=input("숫자를 입력하세요: ") #입력 받으면 문자로 입력 받아짐
print(x, type(x))
if int(x)%2==0: #x는 문자이므로 int()로 형변환해주기!!
    print(f"입력한 숫자{x}는 짝수입니다")
else:
    print(f"입력한 숫자{x}는 홀수입니다")
````

- if문의 조건을 여러 개 주기

````python
# 숫자를 입력 받아서 숫자가 1-10 사이의 숫자인지 판별하고
# 홀수인지 짝수인지 판별해서
# 입력받은 숫자 {x}는 1-10 사이의 숫자이고/숫자가 아니고 홀수/짝수입니다.
x=input("숫자를 입력하세요:")
x=int(x)
#x가 1-10 사이의 숫자이고 짝수인 경우
if 1<=x<=10 and x%2==0:
    print(f"{x}는 1-10 사이의 숫자이면서 짝수입니다")
else:
    print(f"{x}는 1-10 사이의 숫자가 아니거나 홀수입니다")
    
````

## 2. elif : 다중 조건문 사용하기
``` python
if 조건:
    조건이 참일 때 실행할 코드
elif 조건2:
    조건2가 참일 때 실행할 코드
elif 조건3:
    조건2가 참일 때 실행할 코드
else:
    위의 조건1,2,3이 모두 거짓일 때 실행할 코드
```

````python
x=input("숫자를 입력하세요:")
x=int(x)
#x가 1-10 사이의 숫자이고 짝수인 경우
if 1<=x<=10 and x%2==0:
    print(f"{x}는 1-10 사이의 숫자이면서 짝수입니다")
elif 1<=x<=10 and x%2==1:
    print(f"{x}는 1-10 사이의 숫자이면서 홀수입니다")
else:
    print(f"{x}는 1-10 사이의 숫자가 아닙니다")
    
#else, elif 필수는 아님, else 없이 elif만 써도 됨
````

- 점수를 입력 받아서 81-100은 A등급, 61-80은 B등급, 41-60은 C등급, 21-40은 D등급, 0-20은 F등급입니다
- 출력하는 조건문을 만드세요

````python
x=input("점수를 입력하세요:")
x=int(x)
print(x, type(x))
if 81<=x<=100:
    print(f"{x}는 A등급")
elif 61<=x<=80:
    print(f"{x}는 B등급")
elif 41<=x<=60:
    print(f"{x}는 C등급")
elif 21<=x<=40:
    print(f"{x}는 D등급")
elif 0<=x<=20:
    print(f"{x}는 E등급")
````

- 입력값이 0-100 사이인지 판별하고 범위를 벗어나면 0-100 사이의 숫자를 다시 입력하세요

````python
x=input("숫자를 입력하세요:")
x=int(x)
print(x, type(x))
if 0<=x<=100:
    if 81<=x<=100:
        print("A")
    elif 61<=x<=80:
        print("B")
    elif 41<=x<=60:
        print("C")
    elif 21<=x<=40:
        print("D")
    elif 0<=x<=20:
        print("E")
else:
    print(f"""현재 입력된 값은 {x}입니다
    0에서 100 사이의 숫자를 다시 입력하세요""")
````

- 1-50 사이의 숫자를 입력 받아서 3의 배수면 -> 3의 배수입니다.
- 3의 배수가 아니면 -> 3의 배수가 아닙니다

    - 입력 받기, 정수로 변환하기
    - 1-50 사이의 숫자인가?
    - 3의 배수인가? %3==0
    - 결과 출력
   

````python
x=input("숫자를 입력하세요:")
print(x, type(x))
x=int(x)
print(x, type(x))

if 1<=x<=50:
    print(f"{x}는 1-50 사이의 숫자이다")
    if x%3==0:
        print(f"{x}는 1-50 사이의 숫자이면서 3의 배수이다")
    else:
        print(f"{x}는 1-50 사이의 숫자이면서 3의 배수가 아니다")
else:
    print(f"{x}는 1-50 사이의 숫자가 아닙니다. 1-50 사이의 숫자로 다시 입력하세요")
````

# 3. 조건 표현식
- 조건이 하나인 간단한 경우에 씀
```
참일 때 실행할 코드 if 조건 else 거짓일 때 실행할 코드
``` -> 한 줄에 한 번에 적기

````python
# 일반적인 if else 문으로 홀수, 짝수 구분하는 코드
num=int(input("숫자를 입력하세요"))
print(num,type(num))
if num%2==0:
    print("짝수")
else:
    print("홀수")
````

````python
# 일반적인 if else 문으로 홀수, 짝수 구분하는 코드
num=int(input("숫자를 입력하세요 : "))
print(num,type(num))
if num%2==0:
    print("짝수")
else:
    print("홀수")
````

````python
# 조건표현식으로 홀수, 짝수 구분하는 코드
num=int(input("숫자를 입력하세요 : "))
print("짝수") if num%2==0 else print("홀수") # 한 줄에 적기
````
