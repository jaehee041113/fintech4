- 1-10까지 숫자 중 홀수면 홀수입나다. 짝수면 짝수입니다를 찍는 프로그램

````python
for i in range(1,11):
    if i%2==0:
        print(f"{i}는 짝수")
    else:
        print(f"{i}는 홀수")
````

````python
result=[]
for i in range(1,11):
    if i%2==0:
        print(f"{i}는 짝수")
        result.append(f"{i}는 짝수")
    else:
        print(f"{i}는 홀수")
        result.append(f"{i}는 홀수")
result
````

````python
#리스트 표현식으로 만들기
result=[f"{i}는 짝수" for i in range(1,11) if i%2==0 else print(f"{i}는 홀수")]
result
````

````python
#else가 없는 경우 / 있는 경우 -> 리스트 표현식 쓰는 방법 다름!
#else가 있는 경우 : if else 를 먼저 써줌
result = [f"{i}는 짝수" if i%2==0 else f"{i}는 홀수" for i in range(1,11)]
result
````

참일 때 if 조건 else 거짓일 때 for ... (조건 표현식이 앞, for문은 뒤에, if else 함께)

## 1) 조건을 이용해 필터링하기 (if만 쓸 때)

- 1-10 중 짝수만 모으기

````python
# 일반 for문 + if문
evens=[]
for num in range(1,11):
    if num%2==0:
        evens.append(num)
evens
````

````python
#리스트 표현식, if문으로 필터링 -> 이 형태 많이 씀!!
evens2=[num for num in range(1,11) if num%2==0]
#저장될 값-> for앞으로, if 조건 -> for 뒤
evens2
````

# 2) 조건에 따라서 값을 바꾸기 (if~else)

- 1-10 중 짝수면 짝수, 홀수면 홀수를 출력

````python
# for + if -> 이 형태 많이 씀!
labels=[]
for num in range(1,11):
    if num %2==0:
        labels.append("짝수")
    else:
        labels.append("홀수")
labels
````

````python
#리스트 표현식
# 조건표현식 + for문 형식 -> 이 형태 많이 씀! 중요!!
labels2=["짝수" if num %2==0 else "홀수" for num in range(1,11)]
labels2
````

# 제네레이터 표현식
- 리스트 표현식과 문법 동일
- [ ]가 아닌 ( )로 만듦
- 1회용임 - 만들고 나서 반복문으로 내용 꺼내면 내용 사라짐
- 오직 순차 접근만 가능, 인덱싱, 슬라이싱 불가
- 데이터가 너무 크거나, 한 번만 쓸 데이터인 경우 사용

````python
# 1-10까지 숫자 중 짝수인 것만 모은 리스트를 제네레이터로
evens_g=[]
for num in range(1,11):
    if num%2==0:
        evens_g.append(num)
evens_g
````

````python
# 제네레이터 형식 -> ()사용
evens_g=(num for num in range(1,11) if num%2==0)
evens_g
````

# 리스트와 제네레이터 크기 비교

````python
!pip install pympler
````

````python
import sys
from pympler import asizeof
import psutil, os

#메모리 사용량 확인 함수
def show_memory(msg):
    process = psutil.Process(os.getpid())
    print(f"{msg}: {process.memory_info().rss / 1024**2:.2f} MB")

#비교 대상 생성
list_var = [x for x in range(1, 1_000_000)]   # 리스트 내포
gen_var  = (x for x in range(1, 1_000_000))   # 제너레이터 표현식

print("==== sys.getsizeof() 결과 ====")
print("리스트:", sys.getsizeof(list_var))
print("제너레이터:", sys.getsizeof(gen_var))

print("\n==== asizeof.asizeof() 결과 (내부 요소 포함) ====")
print("리스트:", asizeof.asizeof(list_var))
print("제너레이터:", asizeof.asizeof(gen_var))

print("\n==== 프로세스 전체 메모리 사용량 비교 ====")
show_memory("리스트 생성 후")
show_memory("제너레이터 생성 후")
````

# 딕셔너리 함수를 이용해 만들기
- setdefault

````python
data #리스트 안에 딕셔너리 있음
````

````python
for item in data[:2]: #2개만 꺼냄
    #print(item)
    for key, value in item.items():
        print(key, value)
````

````python
result3={}

for i in data[:2]:
    #print(item)
    for key, value in i.items():
        #print(key, value)
        result3.setdefault(key,[]) #키 별로 빈 리스트가 값으로 생김
result3
````

````python
result3={} #result3는 키 밸류 둘다 없는 빈 딕셔너리

for i in data[:2]:
    for key, value in i.items():
        result3.setdefault(key,[]).append(value) #빈 리스트에 밸류 값 하나씩 추가하기
        #result3은 빈 딕셔너리임!
        #for문에서 key와 value를 하나씩 꺼내고,
        #setdefault가 result3에 그 key가 처음 등장할 때만 빈 리스트를 밸류로 생성한다
result3
````

- setdefault

item에서 key, value 꺼냄
→ result3에 key 있는지 확인
→ 없으면 key: [] 생성
→ 그 리스트에 value 추가

````python
#딕셔너리 함수(setdefault) 이용 최종!! 중요!!!

result3={}
for item in data:
    for key, value in item.items():
        result3.setdefault(key,[]).append(value)
#print(result3)

df3=pd.DataFrame(result3) #엑셀 형태 (데이터프레임 형태로 바로 변환 가능!)
df3
````
