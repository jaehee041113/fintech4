# 서식지정자, format 메서드, f-string
- 문자열로만 입력해야 하는 상황에서 같은 위치의 문자를 여러 번 바꿔야 할 때 사용
- 문자의 자릿수 맞춤, 위치 정렬에 사용

# 1. 서식지정자 %d, %s, %f

## 1) %s : 문자를 입력할 때 (string)

```python
print("이 수업의 이름은 %s 입니다." % "국어")
```

```python
for name in ['국어','수학','사회','영어']:
    print("수업 이름은 %s 입니다" % name)
```

## 2) %d : 정수를 입력할 때 (string)

```python
print("올해는 %d년입니다." % 2026)
```

```python
print("올해는 %d년입니다." % 2026.5)
```

## 3) %f : 실수 입력 -> 기본 소숫점 6자리로 표시

```python
print("사람의 정상체온은 약 %f도 이다." %36)
```

- 실수를 입력할 때 소숫점 자리 수를 정하고 싶을 때 -> %.자릿수f

```python
print("사람의 정상체온은 약 %.3f도 이다." %36)
```

## 4) %길이s : 서식지정자로 문자열 길이, 정렬

```python
print("%3s" % "01")
```

```python
"%3s" % "01" #앞에 3자리 확보하고 01 쓰기
```

```python
"%5s" % "01" #앞에 5자리 확보하고 01 쓰기
```

```python
print(150000)
```

```python
print("가격은 %10d 입니다" %150000)
print("가격은 %10d 입니다" %1500)
```

## 5) 여러 항목 동시에 놓기

```python
print("오늘은 %4d년 %2d월 %2d일 입니다" % (2026,1,8))
```

```python
print("오늘은 %4d년 %02d월 %02d일 입니다" % (2026,1,8))
```

```python
company=['삼성', '엘지','현대']
stock=[30,40,50]
```

```python
for com, st in zip(company, stock):
    print("%s의 주가는 %d 입니다" % (com,st))
```

## 2. format 메서드 사용하기
- "문자열 {인덱스번호} 문자열 끝".format(값)

```python
print("{}의 주가는 {}입니다.".format('삼성',30))
```

```python
print("{0}의 주가는 {0}입니다.".format('삼성'))
```

```python
print("{0}의 주가는 {1}입니다.".format('삼성',30))
```

```python
for com, st in zip(company, stock):
    print("{}의 주가는 {}입니다.".format(com, st))
```

## f-string으로 값 넣기

```python
print(company)
print(stock)
```

```python
print(f"{company[0]}의 주가는 {stock[0]}입니다") #변수 직접 넣어서 출력
```

```python
for com, st in zip(company, stock):
    print(f"{com}의 주가는 {st}입니다.") #format보다 f-string이 더 간단함
```

## 3. f-string으로 문자열 정렬, 자릿수 지정해서 출력하기

```python
year=2026
month=1
day=8
print(f"오늘은 {year}년 {month}월 {day}일 입니다")
```

```python
# 한 자릿수를 두 자릿수로 만들 때 -> print("오늘은 %4d년 %02d월 %02d일 입니다" % (2026,1,8))
print(f"오늘은 {year}년 {month:02d}월 {day:02d}일 입니다")
```

```python
url = "https://finance.naver.com/sise/sise_trans_style.naver?sosok="
```

```python
for i in range(1,11):
    print(f"{url}{i:02d}")
```
