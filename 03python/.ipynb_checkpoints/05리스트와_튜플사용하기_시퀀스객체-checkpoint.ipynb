## 리스트 확장하기 -> .extend()
- 리스트에 자료 추가하되 요소만!! 추가
- 통채로 들어가는 게 아니라 그 안에 들어있는 요소만 추가!!
- append() 는 통째로 추가

```python
a=[1,2,3]
b=[4,5,6]
```

```python
#append 쓸 경우 -> list가 그대로 추가됨
a.append(b)
a
```

```python
#extend는 리스트가 통채로 그대로 들어가는 게 아니라 요소만 뽑아서 추가됨
a.extend(b)
a
```

# 튜플 tuple ()
- 튜플로 리스트처럼 모든 자료형을 넣어 만들 수 있다
- 순서가 있는 자료형, 인덱싱, 슬라이싱 가능
- 자료의 추가, 삭제 안 됨
- () 안에 자료 넣어서 만듦
- 자료 1개만 넣어서 만들 때는 반드시 (자료, ) 형식이어야 한다 -> 튜플 원소 1개일 때는 꼭 , 해주기!!
- 튜플 생성 시 자료가 2개 이상인 경우 () 생략 가능
- 빈 튜플 생성 가능


```python
t=(1,2,3)
t
```

```python
t[:2]
```

```python
t[::-1]
```

```python
t.sort() #sort는 요소 바로 변경하고 저장돼서 에러남 - 튜플은 변경 안 됨
```

```python
sorted(t) #저장이 바로 안 돼서 sorted는 가능
#내부적으로 리스트로 바뀜!
```

```python
t=sorted(t)
t
```

```python
t=tuple(t)
t
```

```python
sorted(t) #자료형이 자동으로 내부적으로 리스트로 바뀜
```

# 시퀀스 객체 : 문자열, list, tuple
- 순서 있는 자료형
- 인덱싱, 슬라이싱
- 반복문에 넣어서 자료 하나씩 꺼내기 가능
- 시퀀스 객체 + 시퀀스 객체 (더하기) = 합쳐짐
- 시퀀스 객체 * 숫자 = 숫자만큼 반복 복사돼서 늘어남


```python
s="문자열"
l1=[1,2,3]
l2=[4,5,6]
t1=(1,2,3)
t2=(4,5,6)
```

```python
s*2
```

```python
for item in l1:
    print(item)
```

```python
for item in l1[::-1]:
    print(item)
```
