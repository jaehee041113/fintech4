- 누군가가 미리 사용하기 편하게 만들어 놓은 함수, 클래스
- import 명령을 통해서 프로그램 가장 위에서 모듈을 불러와 사용할 수 있음

# 폴더가 있는지 확인하고 없으면 폴더 생성하기 - 중요!

````python
if not os.path.exists("./testdir"):
    os.mkdir("testdir")
    print("testdir을 생성했습니다")
else:
    print("testdir이 이미 있습니다")
````

````python
# 이 패턴 중요!

dir_name='testdir'

if not os.path.exists(dir_name):
    os.mkdir(dir_name)
    print(f"{dir_name}을 생성했습니다")
else:
    print(f"{dir_name}이 이미 있습니다")
````

# 내 모듈 만들기

- Cal 클래스와 add라는 함수를 모듈로 만들고 불러와서 사용하기

````python
import myfunc # 모듈 그대로 불러오기
````

````python
add(3,5) #에러남
````

````python
# myfunc 모듈 안에 있는 add 함수 사용하기
myfunc.add(3,5)
````

````python
sam=myfunc.Cal(3,5) # 지정해주고 myfunc 모듈 안의 Cal 클래스 불러온 후 생성자(3,5)까지 써주기
````

````python
sam.add()
````

````python
sam.sub()
````
