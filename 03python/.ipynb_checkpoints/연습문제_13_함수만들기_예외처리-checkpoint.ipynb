```python
def hello():
    print("Hello, Python!")
    
hello()
```

```python
def greet(name):
    print(f"안녕하세요 {name}님!")
    
greet("김재희")
```

```python
def add(a, b):
    return a+b

add(1,3)
```

```python
#기본값
def power(x, n=2): 
    return x**n

power(5,3)
```

```python
print(power(3))
```

```python
def say(name, msg='안녕하세요'):
    print(f"{msg} {name}!")

say('민수')
```

```python
def sum_and_diff(a, b):
    return (a+b, a-b)

s, d = sum_and_diff(10, 3)
print(s)
print(d)
```

```python
def quot_rem(a, b):
    return a//b, a%b

q, r = quot_rem(17, 5)
print(q, r)
```

```python
def print_positive(n):
    if n<0:
        print("음수는 처리하지 않습니다")
        return
    else:
        print(n)
        
print_positive(7)
```

```python
def rect(w, h):
    print(f"가로={w}, 세로={h}")
    
rect(3, 5)
```

```python
#가변 위치 인수(*args)

def sum_all(*args):
    return sum(args)

sum_all(1, 2, 3, 4)
```

```python
sum_all(8,9,10,21)
```

```python
def max_of(*args):
    return max(args)

max_of(1, 2, 3, 4)
```

```python
# 가변 키워드 인수(**kwargs)
# 함수 안의 변수 (=가변 키워드 인수 kwargs)
# -> 딕셔너리 형태로 입력

def show_settings(**kwargs):
    for k, v in kwargs.items():
        print(f'{k}={v}')
    
show_settings(mode='dev', debug=True)
#함수 안 변수는 딕셔너리로 입력함
```

```python
def build_person(**kwargs):
    return {'name': kwargs.get('name', 'Unknown'),
        'age': kwargs.get('age', 0),
        'city': kwargs.get('city', 'Seoul')}

#딕셔너리 형태로 반환되도록 직접 설정해줌

build_person(name='Alice', age=25)
    
```
