# 파이썬의 자료형(data type)
## 파이썬의 자료형은 동적으로 변화
* int: 정수(소숫점이 없는 숫자)
* float: 실수(소숫점이 있는 숫자)
* str: 문자/문자열
* list: 여러개의 자료가 순서대로 저장된 자료형, 모든 자료형을 넣어 만들 수 있음
* tuple: list와 비슷하나 한 번 만들면 수정이 안됨, 모든 자료형을 넣을 수 있음
* dictionary(dict): key : value 쌍으로 이루어진 자료형, 모든 자료형을 넣을 수 있음
* set : 집합 자료형, 중복 없는 자료 만들 때, 모든 자료형을 넣을 수 있음, 중복 불가 {} 안에 자료 넣어서 만듦

# 리스트  list [ ]
* 리스트는 [정수, 실수, 문자, 리스트, 튜플 , 디셔너리 ] 모두 넣을 수있다.
* 리스트는 여러 자료가 같이 나열된 형태이므로 순서가 있다.
* 인덱싱, 슬라이싱 가능
* 자료의 추가, 삭제가 가능하다.

```python
l1=[1,2.0,'삼',[1,2,3], (4,5,6),{"key":"value"}] #리스트 안에 있는 리스트, 튜플, 딕셔너리 각각 포장된 하나씩으로 봄
l1
```

list는 순서가 있는 자료형

```python
print(l1,type(l1))
```

```python
l1[2]
```

```python
l1[2:5]
```

```python
l1.append("리스트 자료형 추가")
```

```python
l1
```

```python
del l1[5]
```

```python
l1
```

## 리스트는 리스트 내의 자료형을 그대로 유지함

```python
print(l1, type(l1))

for item in l1:
    print(item, type(item)) #반복문 써서 하나씩 빼서 출력
```

# 딕셔너리형(dict) JSON {"key" : "value"}
* key와 value로 이루어진 데이터형
* json 이라고 불리며 웹상에서 데이터를 주고 받을 때 주로 사용
* 딕셔너리의 value에는 모든 자료형이 올 수 있음
* 딕셔너리는 순서가 없는 자료형
* **key를 호출하면 value가 출력됨 -> 인덱스는 없음! 키로 호출하면 밸류 출력됨**

```python
dic1["이름"] #이름 이라는 키로 호출함
```

```python
dic1["가족"] #가족 이라는 키로 값을 호출함
```

# 형변환 함수
* 형변환: data type 다른 data type으로 바꾸는 것

* 정수로 형변환 int()

```python
a=3
b='3'
print(a+b) #숫자+문자 이기 때문에 오류
```

```python
a=3
b='3'
print(a+ int(b)) ##형변환 함수를 이용해 문자를 정수 숫자로 일시적으로 형변환
```

```python
a=3
b='3'
print(a+int(b))
print("a:", type(a), "b:",type(b))
print("a:", type(a), "int(b):",type(int(b)))
```

* 리스트 형변환 list()

# input() : 키보드 입력을 받는 함수
* input("화면에 표시될 메시지")
* 입력 받는 값은 " "으로 구분해서 여러개를 받을 수 있음
* 입력 받는 값은 모두 str 타입의 문자열!! 숫자를 넣어도 문자열로 인식됨

```python
input_name=input("이름을 입력하세요:")
print(input_name)
```

input함수를 이용해서 숫자 2개를 입력 받아 더한 후 출력하세요.

```python
num1=input("숫자1을 입력하세요")
num2=input("숫자2를 입력하세요")
print("num1+num2:",  num1+num2) #잘못됨
```

```python
num1=input("숫자1을 입력하세요") #-> input에 뭘 넣던지 문자열로 받음!!, 숫자 계산하려면 형변환 해주기
num2=input("숫자2를 입력하세요")
print("num1:", num1, type(num1))
print("num2:", num2, type(num2))
print("num1+num2:",  int(num1)+int(num2))
```

* input().split() 으로 여러 데이터를 list 형태로 받기

- split() : 문자열이 공백을 기준으로 나눠져서 list로 바뀜

```python
nums=input("숫자를 5개 공백으로 구분해서 입력하세요:")
print(nums, type(nums))
nums
```

```python
nums=input("숫자를 5개 공백으로 구분해서 입력하세요:"). split()
print(nums, type(nums))
nums
```

```python
nums
```
