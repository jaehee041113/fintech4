# 산술연산자 +, -, *, /, %, //, **

```python
a=10
b=3
print(a+b)
print(a-b)
print(a/b)
print(a%b)
print(a//b)
```

% 는 나머지를 구하는 연산자 (%2 로 0이면 짝수 1이면 홀수)

```python
print(a%b)
```

```python
print(10%2) #짝수
```

```python
print(11%2) #홀수
```

- %로 배수 구하기

```python
#3의 배수
10%3 #나머지 0 아니어서 10은 3의 배수 아님
```

```python
#7의 배수
14%7 #나머지 0 이어서 14는 7의 배수 맞음
```

```python
#1-100까지 숫자 중에서 7의 배수만 출력하세요
for i in range(1,101):
    if i%7==0:
        print(i, end=' ')
```

# 논리 연산자 
* and, or, not

* and는 두 값이 모두 참이어야 참

```python
True and True
```

```python
True and False
```

or 두 값중 하나라도 참이면 참, 두 값 모두가 거짓이어야 거짓

```python
True or False
```

```python
False or False
```

not 값을 반대로 뒤집음 True=>False, False => True

```python
not True
```

```python
not False
```

* and, or, not의 우선순위
* not, and, or순으로 판단

```python
not True and False or not False
#False and False or True
#False or True
```

```python
10==10 and 10!=5
#True and True
```

# 식별연산자 : is, is not
* is, is not (object, data type)이 같은지 비교
* 값 뿐만 아니라 데이터 타입까지 같은지 비교!!

```python
print(a,b)
```

```python
a==b
```

```python
a is b
#데이터 타입 다름
```

```python
d=3
```

```python
a is d
#값과 데이터 타입 둘 다 같음
```

```python
a is not b
```
