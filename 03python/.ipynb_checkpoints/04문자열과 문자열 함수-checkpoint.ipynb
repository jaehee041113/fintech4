# 문자열 함수

## 문자열 바꾸기 -> .replace(찾을 값, 바꿀 값, 반복횟수=-1)
- replace 함수를 사용 후 결과를 재할당 해야 함

```python
h5= "He!!llo Py!thon!"
h5
```

```python
# h5에서 ! 없애주기
#재할당 꼭 해줘야 결과값 저장됨 중요!!
h5 = h5.replace("!","")

#시프트+탭 누르면 코드 설명 보기 가능
```

```python
h5
```

```python
h5
```

```python
#replace 반복 횟수 정할 수 있음
# h5 에서 ! 삭제하는데 2개만 없애기
h5.replace("!","",2)
```

```python
# 재할당 해줘야 저장됨
h5=h5.replace("!","",2)
h5
```

```python
h5_1="He!!llo Py!t$hon!%"
h5_1
```

```python
# !,$,% 모두 다 없애고 싶을 때
# 문자열 함수는 연이어서 사용 가능
h5_1.replace("!","",2).replace("$","").replace("%","")
```

```python
h5_1=h5_1.replace("!","",2).replace("$","").replace("%","")
h5_1
```

# 문자열의 왼쪽/오른쪽 공백 없애기, 특정문자 제거하기
- lstrip() : 왼쪽 공백 삭제
- rstrip() : 오른쪽 공백 삭제
- strip() : 양쪽 공백 삭제

- strip은 각각 끝쪽만 삭제 -> 사이에 공백 없애고 싶으면 replace 쓰기

```python
h8="    공백이 있는 문자열    "

print(h8)
h8
```

```python
h8.lstrip()
```

```python
h9="..., 공백과 특수문자가 같이 있는 문자열  ^^^"
h9
```

```python
h9.replace(".","").replace(",","").replace("^","").strip()
```

```python
h9.lstrip("., ")
```

```python
h9.rstrip(" ^")
```

```python
h9.strip(" .,^")
```

```python
h10="..., 공백과 특수문자$가 같이 있는\t 문자열  ^^^"
h10
```

```python
h10.strip(".,$\t ^")

#문자열 안에 있는 $,\t 같은 특수문자는 strip으로 없앨 수 없음!
# 문자열 안에 있는 문자 없애려면 replace 사용해야 함
```

## method chaining (매소드 체이닝)
- 여러 개의 함수를 연결해서 사용

```python
h10.strip(".,$\t ^").replace("$",'').replace("\t",'')
```

## 문자열 찾기 : find / index / in
- find() : 문자열이 시작되는 인덱스 번호 출력, 문자열 없으면 -1 출력
- index() : 문자열이 없으면 에러 발생
- in : 문자열이 있으면 True 없으면 False

```python
s="나는 파이썬이 좋아요"
```

```python
#파이썬이라는 단어가 시작되는 인덱스 번호
s.find("파이썬")
```

```python
idx=s.find("파이썬")
s[idx:idx+3] #s[3:6]
```

```python
#찾는 단어가 없으면 -1
s.find("자바") #find는 자료 없으면 -1 출력됨. 에러 안 남
```

```python
#단어의 일부도 찾을 수 있음
s.find("이썬")
```

```python
s.index("파이썬")
```

```python
s.index("자바") #인덱스는 자료 없으면 에러 남
```

```python
"파이썬" in s
```

```python
"자바" in s
```
