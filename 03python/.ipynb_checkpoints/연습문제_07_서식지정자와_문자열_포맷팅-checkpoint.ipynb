```python
subject = "국어"
print("이 수업의 이름은 %s 입니다." % subject)
```

```python
subjects = ["국어","수학","사회","영어"]
for s in subjects:
    print("이 수업의 이름은 %s 입니다." % s)
```

```python
year = 2025
print("올해는 %d년 입니다." % year)
```

```python
temp = 36.5
print("사람의 정상 체온은 약 %f도 입니다." % temp)
```

```python
# %.1f로 소수점 1자리
#temp = 36.5를 소수점 1자리로 표시해 출력
print("사람의 정상 체온은 약 %.1f도 입니다." % temp)
```

```python
# %길이s로 자리수 맞추기
# 문자열 "01"을 3칸 폭으로 출력
print("%3s" % "01")
"%3s" % "01"
```

```python
# %10d로 숫자 폭 맞추기
# 아래 두 숫자를 동일한 폭(10칸)으로 맞춰 출력
print("가격은 %10d 입니다." % 150000)
print("가격은 %10d 입니다." % 15000)
```

```python
# 여러 값 동시에 넣기
# 오늘은 2025년 3월 17일 입니다. (월은 폭 2, 연도는 폭 4)
y, m, d = (2025, 3, 17)
print("오늘은 %4d년 %2d월 %2d일 입니다." % (y, m, d))
```

```python
# zip + %s/%d 출력
# 아래 리스트를 zip으로 묶어 한 줄씩 출력
company = ["삼성전자","LG전자","현대차","대한항공","하나투어"]
stock = [71800,121000,195000,31900,86600]

for com, st in zip(company, stock):
    print("%s의 주가는 %d 입니다." % (com, st))
```
