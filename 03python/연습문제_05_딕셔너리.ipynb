```python
student={"name": "Bob",
 "grades": {"math": 85, "science": 90, "history": 80}}
student
```

```python
student['grades']["science"] 
```

```python
# student의 "grades"에 "english": 95를 추가하고 출력
student['grades'].update(english=95)
student
```

```python
#student의 "math" 점수를 88로 변경하고 출력
student['grades'].update(math=88)
student
```

```python
#student에서 "history" 점수를 삭제하고 출력
del student['grades']['history']
student
```

```python
#딕셔너리 student에서 setdefault를 이용하여 "total"이란 키를 만들고 math, science, english의 합계를 입력하세요.
#단, 합계를 구할 때 값을 수동입력 하지 말고 딕셔너리에서 추출해서 더하도록 하세요.
#.setdefault : 자료 추가 함수
student.setdefault('total', student['grades']['math'] 
                   + student['grades']['science']
                   + student['grades']['english'])
```

```python
#딕셔너리 student2에서 update를 이용하여 'english' 는 100, 'history'는 80를 입력
student2={'name': 'Bob',
 'grades': {'math': 88, 'science': 90, 'english': 100, 'history': 80},
 'total': 270}
student2['grades'].update(english=100, history=80)
student2
```

```python
# dict() 함수를 사용하여 딕셔너리 생성
# = 사용해서 할당하는 방식
#중첩 dict 이용
courses = dict(programming= dict(python=50, java=30, Cpp=20),
              design=dict(Photoshop=25, Illustrator=15))
courses
```

```python
courses['programming']['python']
```

```python
courses['programming']['Cpp'] = 25
courses
```

```python
courses['design']
```

```python
del courses['design']
courses
```

```python
"programming" in courses 
```

```python
# 딕셔너리 courses에서 "design" 항목을 출력하고
#"design" 이 없다면 design key 없음 이라고 출력
courses.get('design', "design key 없음")
```
