# XML로 가져오기
- xml : html처럼 태그 기반으로 자료를 저장한 포맷
- xml parser를 통해서 str을 xml로 변환하는 작업이 필요
- xml로 변환이 되면 태그 기반으로 자료를 찾아서 정리
- 태그에서 자료를 추출할 때는 beautifulsoup이라는 라이브러리를 이용

### 기능 - 처음 2개 주로 씀
- find(1개 찾을 때), find_all(여러 개 찾을 때) 함수: xml, html에서 태그 기반으로 내용을 찾음
- select(여러 개 찾을 때), select_one(1개 찾을 때) 함수: xml, html에서 css selector기반으로 내용을 찾음
- find_all, select는 해당 태그나 css selector를 가진 부분을 모두 찾아서 list로 반환
- find, select_one은 여러 태그나 css selector 중에서 가장 먼저 나오는 태그/selector를 한 개만 찾아줌
- 찾아온 태그 .name: 태그 이름 반환
- 찾아온 태그 .get_text(): 태그 안쪽의 텍스트 반환

````python
# 데이터 추출하는 코드
# 딕셔너리에서 키, 밸류 넣어준 형태 - json과 동일한 형태가 됨

result={}

for item in soup.find_all('item'): # item 태그 안에서 하나씩 꺼내기
    for tags in item.find_all(recursive=False):
    # 공백 있어서 생기는 None값 없애주기 위해 - 태그만(자료만) 들어있는 것들만 나옴
        result.setdefault(tags.name, []).append(tags.get_text()) # 키, 밸류 형태로 넣어줌

df=pd.DataFrame(result)
df
````

### 이전 수업시간 내용 - json 이용
- 전체 데이터 개수로 page 수 계산 후 전체 데이터 수집

````python
# 이전 수업시간 내용 - json 이용
result={}
page=1
total_pages=1

while True:
    
    url="https://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload=dict(divId="indsSclsCd", key="I21006", numOfRows=1000, pageNo=page, type="json", serviceKey=os.getenv("service_key"))
    r=requests.get(url, params=payload) #치킨집(I21006), 한 페이지에 1000개
    #print(r.url)
    #print(r.status_code)
    response=r.json() # 딕셔너리 형태로 옴
     
    # 전체 페이지 수 구하기
    if total_pages==1: # 처음에만 구하고, 다음부턴 안 구함
        if response['body']['totalCount']%1000==0:
            response['body']['totalCount']//1000
        else:
            total_pages=response['body']['totalCount']//1000 +1
    
    # 현재 수집 페이지 출력    
    print(f"{page}/{total_pages} 수집 중  ", end="\r")
        
    # 종료 조건 주기
    if page>total_pages:
        break
    
    # 가게 정보 추출 - json 파일
    for item in response['body']['items']:
        for key,value in item.items():
            result.setdefault(key,[]).append(value)

    page+=1 # 페이지 증가는 마지막에
    time.sleep(0.5) # 과부하 올 수 있어서 시간 여유 주기
    
df=pd.DataFrame(result)
df
````

### 오늘 수업시간 내용 - xml 이용
- 전체 데이터 개수로 page 수 계산 후 전체 데이터 수집
- 조건 : 컬럼명을 한국어 컬럼명으로 만들기 - columns 태그에서 가져와야 함

````python
result={}
page=1
total_pages=1

while True:
    
    # 종료 조건
    if page > total_pages:
        break

    url="https://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload=dict(divId="indsSclsCd", key="I21006", numOfRows=1000, pageNo=page, type="xml", serviceKey=os.getenv("service_key"))
    r=requests.get(url, params=payload) #1000개
    #print(r.url)
    #print(r.status_code)
    soup=bs(r.content,"xml")

    #한국어 컬럼명 추출하기
    kor_cols=soup.find("columns").get_text().split(",")


    # 전체 페이지 구하기
    if total_pages==1: # 처음에만 구하고, 다음부턴 안 구함
        if int(soup.find("totalCount").get_text())%1000==0:
            total_pages=int(soup.find("totalCount").get_text())//1000
        else:
            total_pages=int(soup.find("totalCount").get_text())//1000 +1
        total_pages


    #데이터 추출해서 result에 모으기
    for item in soup.find_all('item'):
        for kor_col, tags in zip(kor_cols, item.find_all(recursive=False)):
            result.setdefault(kor_col, []).append(tags.get_text())
            
    page+=1
    
    time.sleep(0.5)
    
df=pd.DataFrame(result)
df
````

# 모든 카테고리(업종) 데이터 수집하고 업종 별로 csv 파일 만들고 저장하기

````python
upjong=pd.read_excel("./상가상권정보API/소상공인시장진흥공단_상가(상권)정보_업종분류(2302)_및_연계표_v1.xlsx", header=1)
codes= upjong['소분류코드']
names= upjong['소분류명']
````

````python
for code, name in zip(codes, names):
    #print(code, name)
````

````python
# xml 데이터 추출하는 문제 있으면 이 코드 그대로 쓰고 주소만 수정하면 됨!!

for idx, (code, name) in enumerate(zip(codes[:3], names)):
    #print(code, name)
    
    result={}
    page=1
    total_pages=1

    while True:

        # 종료 조건
        if page > total_pages:
            break

        url="https://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
        payload=dict(divId="indsSclsCd", key=code, numOfRows=1000, pageNo=page, type="xml", serviceKey=os.getenv("service_key"))
        r=requests.get(url, params=payload) #key=code
        #print(r.url)
        #print(r.status_code)
        soup=bs(r.content,"xml") #beauitfulsoup 써야 xml 사용 가능!

        #한국어 컬럼명 추출하기 - 이 조건만 추가됨
        kor_cols=soup.find("columns").get_text().split(",")

        # 전체 페이지 구하기 - 같음
        if total_pages==1: # 처음에만 구하고, 다음부턴 안 구함
            if int(soup.find("totalCount").get_text())%1000==0:
                total_pages=int(soup.find("totalCount").get_text())//1000
            else:
                total_pages=int(soup.find("totalCount").get_text())//1000 +1
            total_pages

        print(f"{idx}/{len(names)-1} 카테고리 데이터 {page}/{total_pages}", end="\r")
        
        
        #데이터 추출해서 result에 모으기 -> 여기 달라짐!!
        for item in soup.find_all('item'):
            for kor_col, tags in zip(kor_cols, item.find_all(recursive=False)):
                result.setdefault(kor_col, []).append(tags.get_text())

        page+=1

        time.sleep(0.5)
        
    # 폴더 자동으로 생성하기
    dir_name="shops_data_xml"
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)
        

    df=pd.DataFrame(result)
    df.to_csv(f"./{dir_name}/{name}_데이터.csv", encoding="utf-8-sig")
````
