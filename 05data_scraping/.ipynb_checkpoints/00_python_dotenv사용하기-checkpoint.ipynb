# python-dotenv 사용하기

````python
import os
from dotenv import load_dotenv #함수 불러옴
````

````python
# .env 파일이 있으면 True, 없으면 False
# .env 파일을 다시 불러올 때는 반드시 kernel을 restart 해야 한다.
load_dotenv()
````
