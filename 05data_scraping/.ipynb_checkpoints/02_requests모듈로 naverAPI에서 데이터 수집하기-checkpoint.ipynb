# 2. requests 사용법
``` python
import requests

url='접속주소url' # url
payload=dict(key=value) # parameter
headers=dict(key=value) # headers
r=requests.get(url, parmas=payload, headers=headers) # 서버에 url+parameter로 요청보내기
print(r.url) # 서버에 요청을 보낸 url 출력
print(r.status_code) # 서버에서 보내준 응답코드 200 정상, 400, 500은 오류
response=r.text # r.content, r,json()
```
- r.text : utf-8로 인코딩해서 보여줌 한글이 잘 보임
- r.content : 서버가 보내준 그대로의 자료
- r.json() : requests 모듈 내부의 json 모듈로 text를 json으로 자동 변환

````python
# urllib 방식 -> 이전 수업시간 방식
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request

client_id = os.getenv("Client_Id")
client_secret = os.getenv("Client_Secret")
encText = urllib.parse.quote("삼성전자")
url = "https://openapi.naver.com/v1/search/blog?query=" + encText 

request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8')) 
else:
    print("Error Code:" + rescode)
````

- requests 방식

````python
# requests 방식
# 한 페이지 수집하는 코드임

keyword='삼성전자'
url = "https://openapi.naver.com/v1/search/blog"
payload= dict(query=keyword, display=100, start=1, sort='sim') #파라미터를 딕셔너리 형태로
headers= {"X-Naver-Client-Id": os.getenv("client_id"),
         "X-Naver-Client-Secret": os.getenv("client_secret")}

r=requests.get(url, params=payload, headers=headers)
print(r.url)
print(r.status_code) #200 오면 정상

data=r.json()

result={}

for item in data['items']:
    for key, value in item.items():
        if key in ('title', 'description'):
            value=text_clean(value)
        result.setdefault(key,[]).append(value)
        
df=pd.DataFrame(result)
df
````

# 최종 코드
- requests 방식 - while문 이용해서 전체 페이지 모으기
- 1100개 자료 수집

### <설명>
네이버 블로그 검색 API 규칙 : 한 번 요청에 최대 100개 (display=100)

start 값은 1 ~ 1000까지만 허용

즉:
1페이지: start=1 → 100개
2페이지: start=101 → 100개
…
마지막: start=901 → 100개 → 최대 약 1000~1100개 근처

-> while문 + start 값을 100씩 늘리면서 요청하는 구조가 필요함

- result = {} : 최종 결과를 다 모아둘 그릇, 나중에 DataFrame으로 바꾸려고 만든 딕셔너리
```python
구조
{
  'title': [제목1, 제목2, 제목3, ...],
  'description': [내용1, 내용2, 내용3, ...],
  'link': [...],
  ...
}
```

- 검색 결과 하나씩 꺼내기
블로그 글 한 개 = item / item은 dict
for key, value in item.items(): -> item 안의 모든 key-value 순회
예 : key = 'title' / value = '<b>삼성전자</b> 주가 분석'

-> title / description만 정제

-> result에 누적 저장 : result.setdefault(key, []).append(value)
key가 이미 있으면 그 리스트에 value 추가 / 없으면 빈 리스트 만들고 value 추가


````python
# 코드 + 설명
# requests 방식 - while문 이용해서 전체 페이지 모으기!
# 1100개 자료 수집

result={}

start_num=1
while True: # 무한 반복 + 조건 종료
    
    if start_num==1001:# start_num가 1001이면 안 됨. 강제로 1000으로 줄임
        start_num-=1
        
    elif start_num>1000:
        break
    
    #print("start_num:", start_num, end='\r')
    
    
    keyword='삼성전자' # 네이버에서 검색할 키워드, 나중에 함수로 만들면 여기만 바꾸면 됨
    url = "https://openapi.naver.com/v1/search/blog" # 블로그 검색 API 엔드포인트
    payload= dict(query=keyword, display=100, start=start_num, sort='sim') # 요청 조건 파라미터
    # 인증 정보(headers)
    headers= {"X-Naver-Client-Id": os.getenv("client_id"),
             "X-Naver-Client-Secret": os.getenv("client_secret")}
    #os.getenv(): 코드에 키 직접 쓰지 말고, 환경변수에서 읽어오는 방식(보안)

    r=requests.get(url, params=payload, headers=headers)
    #이 URL에 이 파라미터랑 이 인증 정보 들고 GET 요청 보내줘
    #print(r.url)
    #print(r.status_code)

    data=r.json() #dict로 변환

    

    for item in data['items']:
        for key, value in item.items():
            if key in ('title', 'description'): 
                value=text_clean(value)
            result.setdefault(key,[]).append(value)
            
    
    start_num+=100
    # 다음 요청은: 1 → 101 → 201 → 301 … 그래서 페이지가 넘어감
    
df=pd.DataFrame(result)
df
````

````python
# 코드만 최종 정리

result={}

start_num=1
while True:
    
    if start_num==1001:
        start_num-=1
        
    elif start_num>1000:
        break
    
    
    keyword='삼성전자'
    url = "https://openapi.naver.com/v1/search/blog"
    payload= dict(query=keyword, display=100, start=start_num, sort='sim')
    headers= {"X-Naver-Client-Id": os.getenv("client_id"),
             "X-Naver-Client-Secret": os.getenv("client_secret")}

    r=requests.get(url, params=payload, headers=headers)
    
    data=r.json()


    for item in data['items']:
        for key, value in item.items():
            if key in ('title', 'description'):
                value=text_clean(value)
            result.setdefault(key,[]).append(value)
            
    
    start_num+=100
    
df=pd.DataFrame(result)
df
````

````python
import os
import time
import requests
import pandas as pd
from myfunc import text_clean
from dotenv import load_dotenv
load_dotenv()
````

````python
# 코드 + 설명

keyword = input("검색할 키워드를 입력하세요: ")

# for + while 중첩 구조

for category in ['blog', 'news', 'book']:
    result = {} # category 바뀔 때마다 초기화 - 카테고리별로 결과를 새로 모아야 하기 때문
    start_num = 1
    while True:

        if start_num == 1001:
            start_num -= 1
        elif start_num > 1000:
            break

        print(category, "start_num: ", start_num)    
        url = f"https://openapi.naver.com/v1/search/{category}"
        payload = dict(query=keyword, display=100, start=start_num, sort='sim') #URL이 f-string으로 바뀜
        headers = {"X-Naver-Client-Id" : os.getenv("Client_Id"), 
                   "X-Naver-Client-Secret" : os.getenv("Client_Secret")}
        r = requests.get(url, params=payload, headers=headers)
    #     print(r.url)
    #     print(r.status_code)
        data = r.json()

        for item in data['items']:
            for key, value in item.items():
                if key in ('title', 'description'):
                    value = text_clean(value)
                result.setdefault(key, []).append(value)

        start_num += 100
            
    df = pd.DataFrame(result)
    
    # data라는 폴더가 있나 확인 -> 없으면 새로 만들고, 있으면 그냥 두기
    
    dir_name = "data"
    if not os.path.exists(dir_name):
        os.mkdir(dir_name) # data 폴더 새로 만들기
        print(f"{dir_name}을 생성했습니다.")
    else:
        print(f"{dir_name}이 이미 있습니다.")
        
        
    df.to_csv(f"./data/{keyword}_{category}_result.csv") # CSV 저장 방식 변경 (파일 3개 생성)
    # data 폴더 안에 파일 이름을 검색어 + 카테고리 이름으로 해서 DataFrame(df)을 CSV 파일로 저장
    
    time.sleep(0.5)
````

````python
import os
import time
import requests
import pandas as pd
from myfunc import text_clean
from dotenv import load_dotenv
load_dotenv()
````

````python
# 코드 + 설명

def naver_search(keyword):
    
    for category in ['blog', 'news', 'book']:
        result = {}
        start_num = 1
        while True:

            if start_num == 1001:
                start_num -= 1
            elif start_num > 1000:
                break

            print(category, "start_num: ", start_num)    
            url = f"https://openapi.naver.com/v1/search/{category}"
            payload = dict(query=keyword, display=100, start=start_num, sort='sim')
            headers = {"X-Naver-Client-Id" : os.getenv("Client_Id"), 
                       "X-Naver-Client-Secret" : os.getenv("Client_Secret")}
            r = requests.get(url, params=payload, headers=headers)
        #     print(r.url)
        #     print(r.status_code)
            data = r.json()

            for item in data['items']:
                for key, value in item.items():
                    if key in ('title', 'description'):
                        value = text_clean(value)
                    result.setdefault(key, []).append(value)

            start_num += 100

        df = pd.DataFrame(result)

        dir_name = "data"
        if not os.path.exists(dir_name):
            os.mkdir(dir_name)
            print(f"{dir_name}을 생성했습니다.")
        else:
            print(f"{dir_name}이 이미 있습니다.")
        df.to_csv(f"./data/{keyword}_{category}_result_fuction.csv", encoding="utf-8-sig")

        time.sleep(0.5)
````

# 키워드와 카테고리를 입력하는대로 검색하고 출력하게 하기

````python
def naver_search(keyword, *categories): #(위치인수 keyword, 가변인수 *categories-입력하는 개수만큼)
    
    for category in categories:
        result = {}
        start_num = 1
        while True:

            if start_num == 1001:
                start_num -= 1
            elif start_num > 1000:
                break

            print(category, "start_num: ", start_num)    
            url = f"https://openapi.naver.com/v1/search/{category}"
            payload = dict(query=keyword, display=100, start=start_num, sort='sim')
            headers = {"X-Naver-Client-Id" : os.getenv("Client_Id"), 
                       "X-Naver-Client-Secret" : os.getenv("Client_Secret")}
            r = requests.get(url, params=payload, headers=headers)
        #     print(r.url)
        #     print(r.status_code)
            data = r.json()

            for item in data['items']:
                for key, value in item.items():
                    if key in ('title', 'description'):
                        value = text_clean(value)
                    result.setdefault(key, []).append(value)

            start_num += 100

        df = pd.DataFrame(result)

        dir_name = "data"
        if not os.path.exists(dir_name):
            os.mkdir(dir_name)
            print(f"{dir_name}을 생성했습니다.")
        else:
            print(f"{dir_name}이 이미 있습니다.")
        df.to_csv(f"./data/{keyword}_{category}_result_fuction.csv", encoding="utf-8-sig")

        time.sleep(0.5)
````
