# 첫 페이지 수집 내용에서 total_count 추출 후 페이지 수 계산하고 전체 데이터 수집하기

```python
# 전체 페이지 수 구하기
if response['body']['totalCount']%1000==0:
    response['body']['totalCount']//1000
else:
    total_pages=response['body']['totalCount']//1000 +1
    
total_pages
```

```python
# while 반복문의 조건 만들기
page=1
while True:
     
    # 전체 페이지 수 구하기
    if response['body']['totalCount']%1000==0:
        response['body']['totalCount']//1000
    else:
        total_pages=response['body']['totalCount']//1000 +1
        
    print(f"{page}/{total_pages} 수집 중  ", end="\r")
        
    # 종료 조건 주기
    if page>total_pages:
        break
        
    page+=1
    
    time.sleep(0.5)
```

# 최종 (위의 셀 모두 합침)

```python
result={}
page=1
total_pages=1

while True:
    
    url="https://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload=dict(divId="indsSclsCd", key="I21006", numOfRows=1000, pageNo=page, type="json", serviceKey=os.getenv("service_key"))
    r=requests.get(url, params=payload) #치킨집(I21006), 한 페이지에 1000개
    #print(r.url)
    #print(r.status_code)
    response=r.json() # 딕셔너리 형태로 옴
     
    # 전체 페이지 수 구하기
    if total_pages==1: # 처음에만 구하고, 다음부턴 안 구함
        if response['body']['totalCount']%1000==0:
            response['body']['totalCount']//1000
        else:
            total_pages=response['body']['totalCount']//1000 +1
    
    # 현재 수집 페이지 출력    
    print(f"{page}/{total_pages} 수집 중  ", end="\r")
        
    # 종료 조건 주기
    if page>total_pages:
        break
    
    # 가게 정보 추출
    for item in response['body']['items']:
        for key,value in item.items():
            result.setdefault(key,[]).append(value)

    page+=1
    time.sleep(0.5)
    
df=pd.DataFrame(result)
df
```

```python
import os
import time
import requests
import pandas as pd
from dotenv import load_dotenv
load_dotenv()
```

```python
for code, name in zip(upjong_list['소분류코드'][:2], upjong_list['소분류명']):
    print(code, name)
```

```python
for idx, (code, name) in enumerate(zip(upjong_list['소분류코드'], upjong_list['소분류명'])):
    result ={}
    page = 1
    total_pages = 1
    while True:

        # 종료 조건 주기
        if page > total_pages:
            break

        url = "https://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
        payload = dict(divId="indsSclsCd", key=code, numOfRows=1000, pageNo=page, type="json", serviceKey=os.getenv("service_key"))
        r = requests.get(url, params=payload)
    #     print(r.url)
    #     print(r.status_code)
        response = r.json()

        # 전체 페이지수 구하기
        if total_pages == 1:
            if response['body']['totalCount'] % 1000 == 0:
                total_pages = response['body']['totalCount'] // 1000 
            else:
                total_pages = response['body']['totalCount'] // 1000 +1

        # 현재 수집 페이지 출력
        print(f"{idx}/{len(upjong_list['소분류코드'])}_{name}_{page}/{total_pages} 수집중    ", end="\r")

        # 가게 정보 추출
        for item in response['body']['items']:
            for key, value in item.items():
                result.setdefault(key, []).append(value)

        page += 1
        time.sleep(0.5)

    df = pd.DataFrame(result)
    
    dir_name = "shops_data"
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)
        print(f"{dir_name}을 생성했습니다.")
    else:
        print(f"{dir_name}이 이미 있습니다.")
    
    df.to_csv(f"./shops_data/{name}_가게리스트.csv")  
```
